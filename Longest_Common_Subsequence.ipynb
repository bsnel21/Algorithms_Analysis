##### Brute Force

```python
def LCSbrute(X,Y,i,j):
    if (i==-1 or j == -1):
        return 0
    elif (X[i]==Y[j]):
        return 1 + LCSbrute(X,Y,i-1,j-1)
    return max(LCSbrute(X,Y,i-1,j), LCSbrute(X,Y,i,j-1))
```

```python
import random
import string
import time
```

```python
X = []
Y = []

for i in range(0,10):
    x = random.choice(string.ascii_lowercase)
    X.append(x)
    y = random.choice(string.ascii_lowercase)
    Y.append(y)
    
i = len(X)-1
j = len(Y)-1
```

```python
start_time = time.time()
print(LCSbrute(X,Y,i,j))
print("--- %s seconds ---" % (time.time() - start_time))
print(X)
print(Y)
```

##### Dynamic Programming


```python
def LCS_length(X,Y,m,n):
    b =[[0 for q in range(m)] for r in range(n)]
    c = [[0 for q in range(m)] for r in range(n)]
    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                c[i][j] = c[i-1][j-1]+1
                b[i][j] = "D"
                #print(c[i])
            elif c[i-1][j] >= c[i][j-1]:
                c[i][j]=c[i-1][j]
                b[i][j]="U"
            else: 
                c[i][j]=c[i][j-1]
                b[i][j]="L"
    return c,b
```

```python
A = []
B = []

for i in range(0,5):
    x = random.choice(string.ascii_lowercase)
    A.append(x)
    y = random.choice(string.ascii_lowercase)
    B.append(y)
    
i = len(A)
j = len(B)
```

```python
print(A)
print(B)
LCS_length(A,B, i,j)
```
